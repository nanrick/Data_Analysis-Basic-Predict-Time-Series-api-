```python
import pandas as pd
import datetime
```

```python
# 수동 다운로드 : http://freesis.kofia.or.kr/
# 익스플로러 창에서 조회 가능(Chrome은 접속 불가)
# 일별 데이터 xls다운로드 후 CSV 수동 변환
week_dt = pd.read_csv('D:\\시세모니터링\\융합분석\\DATA\\예탁금\\증시자금추이.csv', encoding='CP949')
week_dt.iloc[1,:] = week_dt.iloc[1,:].str.replace('\n','_')
week_dt.columns = week_dt.iloc[1,:].values
week_dt = week_dt[(week_dt["투자자예탁금_(장내파생상품 거래예수금제외)"].isnull() == False) & (week_dt["구 분"] != '구 분')]
week_dt = week_dt.rename(columns={'구 분' : 'date'})

week_dt.iloc[:,1] = week_dt.iloc[:,1].str.replace(',','').astype(float)
week_dt.iloc[:,2] = week_dt.iloc[:,2].str.replace(',','').astype(float)
week_dt.iloc[:,3] = week_dt.iloc[:,3].str.replace(',','').astype(float)
week_dt.iloc[:,4] = week_dt.iloc[:,4].str.replace(',','').astype(float)
week_dt.iloc[:,5] = week_dt.iloc[:,5].str.replace(',','').astype(float)
week_dt.iloc[:,6] = week_dt.iloc[:,6].str.replace(',','').astype(float)
week_dt = week_dt.reset_index(drop=True)
week_dt
```

```python
# 이전 예탁금 전체 :
# 1998년 06월 18일 ~ 2020년 06월 17일
# xls 파일 오류로 Pandas로 불러오기 실패
# https://muzbox.tistory.com/1012 의 엑셀 VBA 파일을 통한 수동 일괄합치기
path = r'D:\시세모니터링\융합분석\DATA\예탁금\VB0177\Temp'
workbook =  pd.read_excel(path + "\\" + "Consolidation.xls")
workbook.iloc[2,:] = workbook.iloc[2,:].str.replace('\n','_')
workbook.columns = workbook.iloc[2,:].values
workbook = workbook[workbook["투자자예탁금_(장내파생상품 거래예수금제외)"].isnull() == False]
workbook = workbook.drop_duplicates(["구 분"], keep = False)
workbook = workbook.reset_index(drop=True)
workbook["구 분"] = pd.to_datetime(workbook["구 분"].astype(str).str[:10]).astype(str)
workbook.iloc[:,1] = workbook.iloc[:,1].astype(float)
workbook.iloc[:,2] = workbook.iloc[:,2].astype(float)
workbook.iloc[:,3] = workbook.iloc[:,3].astype(float)
workbook.iloc[:,4] = workbook.iloc[:,4].astype(float)
workbook.iloc[:,5] = workbook.iloc[:,5].astype(float)
workbook.iloc[:,6] = workbook.iloc[:,6].astype(float)

workbook = workbook.sort_values(["구 분"])
workbook = workbook.reset_index(drop=True)
workbook = workbook.rename(columns={'구 분' : 'date'})
workbook
```

```python
workbook_mg = pd.concat([workbook, week_dt], axis=0)
workbook_mg = workbook_mg.sort_values(["date"])
workbook_mg = workbook_mg.drop_duplicates(['date'], keep = 'last')
workbook_mg
```

```python
 for i in range(1, len(workbook_mg.columns[1:])+1):
        df = workbook_mg.iloc[:,[0,i]]
        df.to_csv("D:\\시세모니터링\\융합분석\\DATA\\" + datetime.datetime.now().isoformat()[:10].replace("-","") + "\\"+ df.columns[1].split('\n')[0] +".csv", index=False)
#         df.to_csv("D:\\시세모니터링\\융합분석\\DATA\\" + "20201019" + "\\"+ df.columns[1].split('\n')[0] +".csv", index=False)

        print(df.columns[1].split('\n')[0])
```
