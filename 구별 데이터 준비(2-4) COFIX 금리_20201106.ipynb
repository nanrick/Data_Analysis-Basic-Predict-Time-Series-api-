```python
import pandas as pd
```

```python
# https://portal.kfb.or.kr/fingoods/cofix.php
# 신규취급액기준, 잔액기준, 신잔액기준 COFIX [엑셀다운로드]
cofix_dt = pd.read_csv('D:\\시세모니터링\\융합분석\\DATA\\COFIX 금리\\cofix_rt.csv', encoding='CP949')
cofix_dt.columns = cofix_dt.iloc[0].values
cofix_dt = cofix_dt.drop([0])
cofix_dt = cofix_dt.rename(columns={'대상월' : 'Date'})
cofix_dt["Date"] = pd.to_datetime(cofix_dt["Date"], format="%Y-%m-%d")
cofix_dt
```

```python
import datetime
import timedelta

date_all = pd.date_range("2010-01-01", datetime.datetime.now().isoformat()[:10])
date_all = date_all.to_frame().reset_index()
date_all = date_all.rename(columns={0 : "Date"})
date_all = date_all.drop(columns=["index"])
date_all
```

```python
cofix_dt
```

```python
cofix_dt_all = pd.merge(date_all, cofix_dt, how = 'left', on=["Date"])
# cofix_dt_all[cofix_dt_all["신규취급액기준 COFIX"].isnull() == False]
```

```python
cofix_dt_all["신규취급액기준 COFIX"] = cofix_dt_all["신규취급액기준 COFIX"].interpolate(method = "pad")
cofix_dt_all["잔액기준 COFIX"] = cofix_dt_all["잔액기준 COFIX"].interpolate(method = "pad")
# cofix_dt_all[cofix_dt_all["신규취급액기준 COFIX"].isnull() == True][:50]
cofix_dt_all = cofix_dt_all.drop(columns=["공시일", "신 잔액기준 COFIX"])
cofix_dt_all = cofix_dt_all.rename(columns={'신규취급액기준 COFIX' : 'cofix_nb', '잔액기준 COFIX' : 'cofix_bl'})
```

```python
cofix_dt_all.iloc[:,[0,1]].to_csv("D:\\시세모니터링\\융합분석\\DATA\\" + datetime.datetime.now().isoformat()[:10].replace("-","") + "\\"+ "cofix_nb"+".csv", index= False)
# cofix_dt_all.iloc[:,[0,1]].to_csv("D:\\시세모니터링\\융합분석\\DATA\\" + "20201019" + "\\"+ "cofix_nb"+".csv", index= False)
cofix_dt_all.iloc[:,[0,1]]
```
